### **Multinomial Distribution Python Implementation**

```python
import math
import matplotlib.pyplot as plt

def multinomial_probability(N, y, theta):
    """
    Function to calculate the probability of a multinomial distribution.
    Args:
        N: Total number of trials (e.g., total respondents)
        y: List of counts for each class (e.g., [40, 30, 20, 10])
        theta: List of probabilities for each class (e.g., [0.4, 0.3, 0.2, 0.1])
    Returns:
        The probability value of the multinomial distribution.
    """
    # Calculate the multinomial coefficient
    multinomial_coefficient = math.factorial(N)
    for yi in y:
        multinomial_coefficient //= math.factorial(yi)

    # Calculate the probability
    probability = multinomial_coefficient
    for yc, pc in zip(y, theta):
        probability *= (pc ** yc)

    return probability*100

# Example of a survey
N = 100  # Total number of respondents
theta = [0.4, 0.3, 0.2, 0.1]  # Probabilities for each class
# Some possible y combinations
y_combinations = [
    [40, 30, 20, 10],
    [35, 35, 20, 10],
    [50, 30, 15, 5],
    [25, 25, 30, 20],
    [45, 25, 20, 10]
]

# Calculate probabilities
probabilities = [multinomial_probability(N, y, theta) for y in y_combinations]

# Visualize the results as a bar chart
y_labels = [str(y) for y in y_combinations]  # Convert each y combination to a string for labeling
plt.bar(y_labels, probabilities, color='skyblue')

# Set chart labels
plt.xlabel("y Combinations")
plt.ylabel("Probability [%]")
plt.title("Multinomial Distribution Probabilities [%]")
plt.xticks(rotation=10)
plt.show()
```

$\therefore$ It can be confirmed that the closer **y** aligns with the probability distribution **theta**, the higher the likelihood of occurrence.
