## **Maximum Likelihood Estimation using Moving Average Method.**

This code performs Maximum Likelihood Estimation (MLE) using moving averages on data generated from a normal distribution with a mean of 5 and a variance of 1.

---

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Initialize variables
moving_averages = []  # List to store moving averages over time
moving_average = 0    # Current moving average

true_mean = 5  # True mean of the normal distribution

# Iterate for 2000 time steps
for t in range(2000):
    # Generate a random sample y_t from a normal distribution
    # loc=true_mean sets the mean, scale=1 sets the standard deviation
    y_t = stats.norm.rvs(loc=true_mean, scale=1, size=1, random_state=None)

    if t == 0:  # Handle the initial state
        moving_average = y_t  # Initialize moving average with the first sample
        moving_averages.append(moving_average)
    else:
        # Update the moving average using the formula:
        # moving_average = previous_average + (1/t) * (current_sample - previous_average)
        moving_average = moving_average + 1/t * (y_t - moving_average)
        moving_averages.append(moving_average)

# Plot the results
plt.plot(
    np.array(range(len(moving_averages))),  # X-axis: Time steps
    len(moving_averages) * [true_mean],     # Y-axis: True mean (constant line)
    label="True mean"
)
plt.plot(moving_averages, label="Moving Average")  # Plot moving averages over time

# Add plot details
plt.title("Moving Average for MLE")
plt.xlabel("Time Steps")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

---

### **Conclusion**
$\therefore$  As the time steps increase, the moving average converges to the true mean value.

In the case of the Gaussian function, the mean Î¼ that maximizes the likelihood function corresponds to the sample mean, which is equivalent to the **moving average**
