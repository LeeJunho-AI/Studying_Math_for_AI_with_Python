## **Convolution**

The **probability** of the sum of the rolls of two dice can be calculated using convolution, as the outcomes of the two dice rolls are independent of each other.

---  
### **Formula**
$$
\sum\limits_{k=-\infty}^{\infty}{p(x_1)p(x2)}\\
y=x_1+x_2
$$
y is the sum of rolls of two dice.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the probability distribution of a single die (1 to 6)
dice_distribution = np.array([1/6]*6)

# Perform the convolution to find the probability distribution of the sum of two dice
sum_distribution = np.convolve(dice_distribution, dice_distribution)

# Define the possible sums (2 to 12)
sums = np.arange(2, 13)

# Plot the probability distribution of the sums
plt.figure(figsize=(8, 6))
plt.bar(sums, sum_distribution, color='lightgreen', alpha=0.7, edgecolor='k')
plt.title('Probability Distribution of the Sum of Two Dice')
plt.xlabel('Sum of Dice')
plt.ylabel('Probability')
plt.xticks(sums)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
