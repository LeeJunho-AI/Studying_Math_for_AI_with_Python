### Monte Carlo approximation Implementation


```python
import numpy as np
```

```python
N = 10000 # Sample Number

dice_rolls = np.random.randint(1, 7, size = N)
```

```python
print(dice_rolls.shape)
print(dice_rolls.dtype)
```

```python
unique, counts = np.unique(dice_rolls, return_counts = True)
probabilities = counts/N
print('The probabilities for each face of the die')
for i, prob in enumerate(probabilities):
  print(f'{i + 1}: {prob}')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(unique, probabilities, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Dice Face')
plt.ylabel('Probability')
plt.title('Histogram of Dice Rolls')
plt.xticks(unique)
plt.ylim(0, max(probabilities) + 0.02)
plt.show()
```

$\therefore$ Since the probability of each face appearing is approximately 1/6, it can be seen that the actual probability mass is almost identical.
