### **Binomial Distribution**
---
##### **Explain**
Player is about to play a lottery with a 70% chance of winning.  

Calculate the probabilities of winning exactly 1 time up to 10 times out of 10 consecutive trials  
<br>  

---
##### **Formula**
$$
Bin(s|N,\theta) = \displaystyle \frac{N!}{(N-s)!s!}\theta^s(1-\theta)^{N-s}
$$

s: Number of Successes  
N: Number of Trials  
$\theta$: Success Probability  


```python
import math
import matplotlib.pyplot as plt

def bin_caculator(s, N, theta):
  output = math.factorial(N) / (math.factorial(N-s) * math.factorial(s)) * (theta**s) * ((1-theta)**(N-s))
  return output*100

N = 10
theta = 0.7
results = []

for s in range(1, N+1): # for N starts at 1 and going up to 10
  results.append(bin_caculator(s, N, theta))

plt.bar(list(range(1, 11)), results)
plt.xticks(ticks=list(range(1, 11)), labels = list(range(1, 11)))
plt.title('Binomial Distribution')
plt.ylabel('Probability (%)')
plt.xlabel('Number of Successes')
plt.show()
```
